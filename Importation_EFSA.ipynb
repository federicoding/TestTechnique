Première chose à faire, séparer la première colonne en deux.

```python
df = pd.read_csv(filepath_or_buffer=url, sep='\t', header=0)
```

```python
def get_prem(element):
  return element.split(',')[0]
```

```python
def get_second(element):
  return element.split(',')[1]
```

```python
df['indic_ur'] = df['indic_ur,cities\\time'].apply(get_prem)
```

```python
df['cities'] = df['indic_ur,cities\\time'].apply(get_second)
```

###Avec strip

```python
df.columns = df.columns.str.strip()
```

Avec un peu de code, pour être méthodique

```python
def whitespace_remover(dataframe):
  #iterations over columns
  for i in dataframe.columns:
    #check le type de chaque colonne
    if dataframe[i].dtype == 'object':
      #appliquer le strip
      dataframe[i] = dataframe[i].map(str.strip)
    else:
      #si rien, ben rien
      pass
```

```python
whitespace_remover(df)
```

```python
print(df)
```
